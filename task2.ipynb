## 1)which are the highest grossing movies with data

```python
import pandas as pd
import matplotlib.pyplot as plt
top_grossing = df2.sort_values(by='worlwide_gross_income', ascending=False).head(10)
print("Top 10 Highest Grossing Movies:")
display(top_grossing[['title', 'worlwide_gross_income']])
plt.scatter(top_grossing['title'], top_grossing['worlwide_gross_income'])
plt.xticks(rotation=90)
plt.xlabel("Movie Title")
plt.ylabel("Worldwide Gross Income")
plt.title("Top 10 Highest Grossing Movies")
plt.tight_layout()
plt.show()
```

## 5) what is gross income per director

```python
import pandas as pd
import matplotlib.pyplot as plt
df2['worlwide_gross_income'] = pd.to_numeric(df2['worlwide_gross_income'], errors='coerce')
director_income = df2.groupby('director')['worlwide_gross_income'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
director_income.head(10).plot(kind='barh', color='teal', edgecolor='black')
plt.title('Top 10 Directors by Worldwide Gross Income')
plt.ylabel('Total Gross Income')
plt.xlabel('Director')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
