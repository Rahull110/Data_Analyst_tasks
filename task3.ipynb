## 9.Write a program to show the Monthly Release Patterns and Viewership Hours (2023) using bar and scatter plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\rahul\OneDrive\Desktop\DataAnalytics\netfix.csv")

df["Viewership_Hours"] = df["Viewership_Hours"].astype(str).str.replace(",", "").astype(float)
df["Release_Month"] = pd.to_datetime(df["Release_Date"], errors="coerce").dt.month

# Monthly counts and hours
monthly_counts = df["Release_Month"].value_counts().sort_index()
monthly_hours = df.groupby("Release_Month")["Viewership_Hours"].sum()

plt.figure(figsize=(10,6))

# --- BAR (Primary y-axis) ---
ax1 = plt.gca()
ax1.bar(monthly_counts.index, monthly_counts.values, alpha=0.6, label="Number of Releases",color="yellow")
ax1.set_xlabel("Release Month")
ax1.set_ylabel("Number of Releases", color="black")

# --- SCATTER (Secondary y-axis) ---
ax2 = ax1.twinx()
ax2.scatter(monthly_hours.index, monthly_hours.values, s=100, color="red", label="Viewership Hours")
ax2.plot(monthly_hours.index, monthly_hours.values, color="black", alpha=0.7)
ax2.set_ylabel("Viewership Hours", color="orange")

plt.title("Monthly Release Patterns and Viewership Hours (2023)")

plt.tight_layout()
plt.show()
```

## 10)  write a program to show the weekly relese patterns and vivewership hours (2023) using bar and scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt

df["Viewership_Hours"] = df["Viewership_Hours"].astype(str).str.replace(",", "").astype(float)
df["Release_Date"] = pd.to_datetime(df["Release_Date"], errors="coerce")
df["Weekday"] = df["Release_Date"].dt.day_name()

# Order of days
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Counts and hours
weekday_counts = df["Weekday"].value_counts().reindex(days)
weekday_hours = df.groupby("Weekday")["Viewership_Hours"].sum().reindex(days)

plt.figure(figsize=(10,6))

# --- BAR (primary y-axis) ---
ax1 = plt.gca()
ax1.bar(days, weekday_counts.values, alpha=0.6, label="Releases", color="orange")
ax1.set_ylabel("Number of Releases", color="blue")

# --- SCATTER (secondary y-axis) ---
ax2 = ax1.twinx()
ax2.scatter(days, weekday_hours.values, s=120, color="red", label="Viewership Hours")
ax2.plot(days, weekday_hours.values, color="black", label="Viewership Hours")
ax2.set_ylabel("Viewership Hours", color="darkblue")

plt.title("Weekly Release Patterns and Viewership Hours (2023)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 11) Write a program to show the movies,series released on holidays such as new year,independence day, valentines day,christmas etc.

```python
import pandas as pd


# Convert release date to datetime
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')

# Extract month and day
df['Month'] = df['Release_Date'].dt.month
df['Day'] = df['Release_Date'].dt.day

# Define major holidays
holidays = {
    "New Year": [(1, 1)],
    "Valentine's Day": [(2, 14)],
    "Independence Day (India)": [(8, 15)],
    "Christmas": [(12, 25)]
}

# Find releases on holidays
holiday_releases = []

for holiday, dates in holidays.items():
    for month, day in dates:
        matches = df[(df['Month'] == month) & (df['Day'] == day)]
        if not matches.empty:
            temp = matches.copy()
            temp["Holiday"] = holiday
            holiday_releases.append(temp)

# Combine results
if holiday_releases:
    result = pd.concat(holiday_releases)
    print(result[['Title', 'Content_Type', 'Release_Date', 'Holiday']])
else:
    print("No holiday releases found in the dataset.")
```
