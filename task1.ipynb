## 1)convert gross and budget to million

```python

df1["Gross"]=df1["Gross"]/1000000
df1["budget"]=df1["budget"]/1000000
print(df1[["Gross","budget"]])

```

## 2)create a new coloum profit

```python

df1["profit"]=df1["Gross"]-df1["budget"]
df1[["budget","Gross","profit"]]
```

## 3)sort the data

```python
df1=df1.sort_values(by='profit',ascending=False)
print(df1[['profit']])
```

## 4)get the top 10 profit movie

```python
top10_profit_movies=df1.iloc[0:10]
print(top10_profit_movies[["Title","profit"]])
```

## 5)find movies with negative profit

```python
negative_profit_movies=df1[df1["profit"]<0]
print(negative_profit_movies[["Title","profit"]])
```

## 6)change the scale of metacritic

```python
df1['MetaCritic']=df1['MetaCritic']/10
print(df1)
```

## 7)find the average ratings

```python
import pandas as pd
import matplotlib.pyplot as plt  
df1 = pd.DataFrame({
    'Movie': ['Inception', 'Interstellar', 'The Dark Knight', 'Tenet'],
    'IMDb_rating': [8.8, 8.6, 9.0, 7.4],
    'MetaCritic': [74, 74, 84, 69]
})
df1['Avgrating'] = df1['IMDb_rating'] + df1['MetaCritic'] / 2

print(df1)
plt.figure(figsize=(8, 5))
plt.barh(df1['Movie'], df1['Avgrating'])
plt.title('Average Rating (IMDb + MetaCritic/2)')
plt.xlabel('Movie')
plt.ylabel('Average Rating')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()  
```

## 8)sort in descending order of avg rating

```python
df1=df1.sort_values(by='Avgrating',ascending=False)
print(df1[["Title","Avgrating"]])
```

## 9) find the movie with metacritic <0.5>and Imdb_ratings>8

```python
movie=df1[(df1['MetaCritic']<0.5)&(df1['IMDb_rating']>8)]
print(movie[['Title','MetaCritic','IMDb_rating']].notna().sum())
```

## 10)find top 5 actor trios with highest fb likes

```python
import pandas as pd
import matplotlib.pyplot as plt
df1['total_actor_fb_likes'] = (
    df1['actor_1_facebook_likes'] +
    df1['actor_2_facebook_likes'] +
    df1['actor_3_facebook_likes']
)
df1 = df1[['actor_1_name','actor_2_name','actor_3_name','total_actor_fb_likes']]
df1 = df1.sort_values(by='total_actor_fb_likes', ascending=False).head(5)
df1['actor_trio'] = df1['actor_1_name'] + ' | ' + df1['actor_2_name'] + ' | ' + df1['actor_3_name']
print(df1[['actor_trio','total_actor_fb_likes']])
plt.figure(figsize=(10,6))
plt.barh(df1['actor_trio'], df1['total_actor_fb_likes'])
plt.xlabel('Total Facebook Likes')
plt.ylabel('Actor Trios')
plt.title('Top 5 Actor Trios with Highest Facebook Likes')
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.show()
```
